```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import os
import sklearn
from sklearn.preprocessing import StandardScaler
```

```python
path = r'C:\Users\kuohe\ClimateWins\02 Data'
```

```python
# Import the data set
df_climate = pd.read_csv(os.path.join(path, 'Processed Weather Predictions.csv'))
```

```python
df_climate.head()
```

```python
df_climate.reset_index(inplace=True)
df_climate.rename(columns={'index':'id'}, inplace=True)
df_climate.head()
```

```python
#Checking for null values
df_climate.isnull().sum()
```

```python
# Drop NA values
df_climate.dropna(inplace=True)
```

```python
df_climate.shape
```

```python
# Checking for duplicates
df_climate_dup = df_climate[df_climate.duplicated()]
```

```python
df_climate_dup
```

```python
# Setting up argument for column without 'DATE'
columns = df_climate.columns
cols_no_date = [x for x in columns if "DATE" not in x]

date_column = df_climate['DATE']

scaler = StandardScaler()
df_climate_scaled = pd.DataFrame(scaler.fit_transform(df_climate[cols_no_date]), columns=cols_no_date)

df_climate_scaled['DATE'] = date_column
```

```python
df_climate_scaled.head()
```

```python
# Saving the correctly scaled dataframe
df_climate_scaled.to_pickle(os.path.join(path, 'Dataset_scaled_correct.pkl'))
```
